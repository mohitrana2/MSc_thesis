```python
import pandas as pd
# Define file paths
data_matrix_path = '/Users/mohitrana/Documents/final_project/preprocessing/L1000-FWD/output/data_matrix.csv'
InChIKey_path = '/Users/mohitrana/Documents/final_project/preprocessing/adb1019/final_data_matrix/BRD_InChIKey.txt'
ATC_codes_path = '/Users/mohitrana/Documents/final_project/preprocessing/adb1019/final_data_matrix/drugbank_results_ATC.csv'
output_file_path = '/Users/mohitrana/Documents/final_project/preprocessing/L1000-FWD/output/data_matrix_InChIKey_ATC.csv'
```

```python
# Read the first file into a DataFrame
df1 = pd.read_csv(data_matrix_path, sep=',')
# Extract the string 'BRD-' followed by 9 alphanumeric characters
df1['BRD_Code'] = df1['rid'].str.extract(r'(BRD-\w{9})')
print("First DataFrame (Comma-separated):")
#print(df1.head())
df1
```

```python
# Read the second file into a DataFrame
df2 = pd.read_csv(InChIKey_path, sep='\t')
print("\nSecond DataFrame (Comma-separated):")
#print(df2.head())
df2
```

```python
# Merge the DataFrames on the "BRD_Code" column
merged1_df = pd.merge(df1, df2, on='BRD_Code', how='inner')
print("\nMerged DataFrame:")
#print(merged1_df.head())
merged1_df
```

```python
# Read the second file into a DataFrame
df3 = pd.read_csv(ATC_codes_path, sep=',', quotechar='"')
print("\nThird DataFrame (Comma-separated with string delimiters):")
print(df3.head())
#df3
```

```python
# Merge the DataFrames on the "InChI Key" column
merged2_df = pd.merge(merged1_df, df3, on='InChI Key', how='inner')
print("\nMerged DataFrame:")
#print(merged2_df.head())
merged2_df
```

```python
merged2_df.drop(columns=['Withdrawn', 'Investigational','Synonyms','Approved','Experimental','Category','Name','Small molecule','InChI Key'], inplace=True)
print(merged2_df.head())
```
