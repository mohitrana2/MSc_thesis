Now, we have to check how many of small molecules are unique in our dataset. To do this, we first need to 

i) drop those molecules that have no ICD-11 labels and then,

ii) extract the BRD Codes from the first column and add that as the last column of our dataframe. Then, we can use `unique()` to find out how many of them are infact unique!

#### using dropna method in pandas

```python
last_column_name = mapped_file.columns[-1]
df_cleaned = mapped_file.dropna(subset=[last_column_name])
```

```python
# Recheck for null entries in the cleaned dataframe
df_cleaned.isnull().sum()
```

```python
import re

# Define a function to extract the BRD- ID
def extract_brd_id(rid):
    match = re.search(r'BRD-[A-Za-z0-9]{9}', rid)
    if match:
        return match.group(0)
    return None

# Apply the function to the 'rid' column and create a new column 'BRD_id'
df_cleaned['BRD'] = df_cleaned['rid'].apply(extract_brd_id)
```

```python
# now check how many of small molecules are infact unique..
unique_count = df_cleaned['BRD'].nunique()
print(f"We have {unique_count} unique small molecules in the dataframe.")
```
