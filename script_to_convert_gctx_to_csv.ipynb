```python
import pandas as pd
from cmapPy.pandasGEXpress import parse

# Path to your .gctx file
gctx_file_path = 'L1000-FWD/CD_signatures_LM_42809x978.gctx'

# Read the GCTX file
gctoo = parse.parse(gctx_file_path)

# Extract the data matrix as a pandas DataFrame
data_df = pd.DataFrame(data=gctoo.data_df.values, index=gctoo.data_df.index, columns=gctoo.data_df.columns)

# Display the original data DataFrame (optional)
print("Original Data Matrix:")
print(data_df.head())

# Extract row and column metadata
row_metadata_df = gctoo.row_metadata_df
col_metadata_df = gctoo.col_metadata_df

# Display the row and column metadata DataFrames (optional)
print("\nRow Metadata:")
print(row_metadata_df.head())
print("\nColumn Metadata:")
print(col_metadata_df.head())

# Optionally, save the main data matrix to a CSV file
output_file_path = 'L1000-FWD/data_matrix.csv'
data_df.to_csv(output_file_path)

# Optionally, save the row and column metadata to CSV files
row_metadata_output_path = 'L1000-FWD/row_metadata.csv'
col_metadata_output_path = 'L1000-FWD/col_metadata.csv'
row_metadata_df.to_csv(row_metadata_output_path)
col_metadata_df.to_csv(col_metadata_output_path)
```
