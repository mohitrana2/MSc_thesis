```python
import requests
import time
import os
#from google.colab import drive

# Mount Google Drive
#drive.mount('/content/drive')

def get_cid_from_inchi(inchi_key, retries=5, backoff_factor=1):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/inchikey"
    url = f"{base_url}/{inchi_key}/cids/JSON"
    for attempt in range(retries):
        try:
            response = requests.get(url, timeout=10)  # Set a timeout for the request
            response.raise_for_status()  # Raise an error for bad status codes
            data = response.json()
            if "IdentifierList" in data and "CID" in data["IdentifierList"]:
                return data["IdentifierList"]["CID"]
        except (requests.exceptions.ConnectTimeout, requests.exceptions.ReadTimeout, requests.exceptions.RequestException) as e:
            print(f"Error fetching CID for InChI key {inchi_key}: {e}")
            wait_time = backoff_factor * (2 ** attempt)
            print(f"Retrying in {wait_time} seconds...")
            time.sleep(wait_time)
    return None

def get_cids_for_inchi_keys(inchi_keys, log_file_path, output_file_path):
    cids = {}
    processed_keys = set()

    if os.path.exists(log_file_path):
        with open(log_file_path, 'r') as log_file:
            for line in log_file:
                key, cid = line.strip().split(' -> ')
                cids[key] = cid
                processed_keys.add(key)

    with open(log_file_path, 'a') as log_file, open(output_file_path, 'a') as output_file:
        for inchi_key in inchi_keys:
            if inchi_key in processed_keys:
                continue
            cid = get_cid_from_inchi(inchi_key)
            if cid:
                cids[inchi_key] = cid
                log_file.write(f"InChI Key: {inchi_key} -> CID: {cid}\n")
                log_file.flush()
                output_file.write(f"InChI Key: {inchi_key} -> CID: {cid}\n")
                output_file.flush()
            else:
                log_file.write(f"InChI Key: {inchi_key} -> Error: Not found\n")
                log_file.flush()
            time.sleep(0.5)
    return cids

def read_inchi_keys_from_file(file_path):
    inchi_keys = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()  # Adjust the delimiter if needed
            if parts:
                inchi_key = parts[0]  # Assuming the InChI key is the first column
                inchi_keys.append(inchi_key)
    return inchi_keys

# Path to the .txt file containing the InChI keys
file_path = '/home/localuser/Documents/mohit/inchi-to-cid/InChIKey-to-CID_remaining.csv'
# Path to the log file to save progress
log_file_path = '/home/localuser/Documents/mohit/progress_log.txt'
# Path to the output file
output_file_path = '/home/localuser/Documents/mohit/inchi-to-cid/remaining_output.txt'

# Read the InChI keys from the file
inchi_keys = read_inchi_keys_from_file(file_path)

# Get CIDs for the InChI keys with progress logging
cids = get_cids_for_inchi_keys(inchi_keys, log_file_path, output_file_path)

print(f"Results written to {output_file_path}")
```
